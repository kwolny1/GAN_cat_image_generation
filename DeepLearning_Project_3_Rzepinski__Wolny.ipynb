```python
import os, shutil

all_images = sorted(os.listdir("./cat-dataset/cats/Data/"))
train_imgs = all_images[:10000]
eval_imgs = all_images[1000:2000]

os.makedirs("data_path", exist_ok=True)
os.makedirs("eval_path", exist_ok=True)

for f in train_imgs:
    shutil.copy(f"./cat-dataset/cats/Data/{f}", "data_path/")

for f in eval_imgs:
    shutil.copy(f"./cat-dataset/cats/Data/{f}", "eval_path/")
```
